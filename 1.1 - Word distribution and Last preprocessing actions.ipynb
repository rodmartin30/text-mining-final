# Remove specifict words

```python
# Getting data.
data, original_data = load_corpus()
removed = {
    'omitted':'',
    'jeje':'',
    'ero':'',
    'vos':'',
    '..':'',
    '...':'',
    '....':'',
    '.....':'',
    'media':'',
    'location':'',
    'missed':'',
    'res':'',
}
# Remove words which appears more than 50% of times.
for word in distribution_words:
    if min(len(distribution_words),len(data)) * 0.4 < distribution_words[word]:
        print(word)
        removed[word] = ''
new_data = []
for phrase in data:
    new_phrase = []
    for word in phrase:
        if word in removed:
            continue
        new_phrase.append(word)
    new_data.append(new_phrase)
data=new_data.copy()
```
